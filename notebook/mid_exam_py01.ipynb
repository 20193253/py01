### Q5) 다음 bool 값 출력 중 정확한 것은 ?




````
# bool 자료형
# 문자열, list, tuple, dictionary, set : 데이터 타입이 null일 경우 False 
# int : 0 = False / 1 = True

print(bool([0, 1, 2])) # 값이 존재하는 리스트 : True
print(bool([0])) # 값이 존재하는 리스트  : True
print(bool((0))) # null set : False
print(bool((0,))) # 값이 존재하는 튜플 # (0, ) : 요소가 '0' 하나인 튜플 : True
# bool(0) # False
````

### Q8) 다음의 결과가 출력되지 않는 것은 ?

```
'I ate 10 apples. so I was sick for three days.'

number = 10
day = "three"

"I ate {0} apples. so I was sick for {1} days.".format(10, day)
f"I ate {number} apples. so I was sick for {day} days."
f"I ate {number} apples. so I was sick for {'three'} days."
"I ate {0} apples. so I was sick for {1} days.".format(number = 10, day)
```



````
number = 10 
day = "three"

# 문자열 포매팅 : format()

# 인덱스 사용

"I ate {0} apples. so I was sick for {1} days.".format(number, day)
"I ate {0} apples. so I was sick for {1} days.".format(10, "three")
"I ate {0} apples. so I was sick for {1} days.".format(number, "three")
"I ate {0} apples. so I was sick for {1} days.".format(10, day)

# 값을 가진 변수 사용

"I ate {number} apples. so I was sick for {day} days.".format(number = 10, day = "three")

# 인덱스와 변수를 혼용하여 사용

"I ate {0} apples. so I was sick for {day} days.".format(number, day = "three")
````

````
# 문자열 포매팅 : f

f"I ate {number} apples. so I was sick for {day} days."
f"I ate {number} apples. so I was sick for {'three'} days."
f"I ate {10} apples. so I was sick for {day} days."
````

````
# 에러코드 

"I ate {0} apples. so I was sick for {1} days.".format(number = 10, day) 
# 'number = 10' : 'number' 라는 변수에 값을 부여하였으나 문자열에서는 인덱스 {0} 을 사용하였다.

# {0} --> {number}

# "I ate {number} apples. so I was sick for {0} days.".format(number = 10, day)

# positional argument follows keyword argument

# {number} <- 'number = 10' : keyword argument, 키워드 인수 방식 - 인수에 키워드를 부여하여 대입 
# {0} <- day : positional argument, 위치 인수 방식 - 정해진 인수의 위치에 값을 대입 
# 이를 혼용할 경우 위치 인수를 처리한 후 키워드 인수를 처리하는데, format(number = 10, day)에서는 키워드 인수를 먼저 제시하였으므로 에러가 발생한다. 


# 키워드 인수와 위치 인수의 순서를 바꾸면 올바르게 실행된다. 
#"I ate {0} apples. so I was sick for {day} days.".format(number, day = "three")
````

### Q9) 리스트 내포를 사용해 1부터 1000 이하의 자연수 중 5의 배수의 합을 구하는 한 줄 코드를 작성하시오.

````
# for 문을 사용하여 작성

total = 0

for num in range(1, 1001) :
  if num % 5 == 0 :
    total = total + num

print(total)
````

````
# 위의 코드를 리스트 내포를 사용하여 한 줄 코드로 작성

sum(num for num in range(1, 1001) if num % 5 == 0)
````

### Q10) 4개 과목의 평균 성적을 구하시오. [a], [b]에 적합한 코드는 ?

```
subjects = ['python', 'english', 'math', 'science']
scores = [88, 77, 66, 95]
total = 0

for subject, score in [a]________(subjects, scores) :
   total += score
   print(subject, score)

print("평균 점수 = {0}".format([b]________))
```



for 문과 iterator를 사용한 과목명과 점수 출력 및 평균 성적 구하기

````
# zip()

subjects = ['python', 'english', 'math', 'science']
scores = [88, 77, 66, 95]
total = 0

for subject, score in zip(subjects, scores) :
  total += score
  print(subject, score)

print("평균 점수 = {0}".format(total / len(subjects)))
````

````
# enumerate()

total = 0

for index, subject in enumerate(subjects) :
  subject = subjects[index]
  score = scores[index]
  total += score
  print(subject, score)

print("평균 점수 = {0}".format(total / len(subjects)))
````

````
# range()

total = 0

for index in range(len(subjects)) :
  subject = subjects[index]
  score = scores[index]
  total += score
  print(subject, score)

print("평균 점수  = {0}".format(total / len(subjects)))
````

문제는 zip() 코드 형식에 해당한다. 

[a] = zip
[b] = total / len(subjects)
